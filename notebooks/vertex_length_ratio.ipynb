```python
import sys; sys.path.insert(0, '..')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import trajectory_reconstruction_tradeoff as T
```

```python
df = pd.read_csv('../comp_vertex_length.csv')
df
```

```python
df.columns
```

```python
datasets = df['dataset'].unique()
for dataset in datasets:
    sdf = df[df['dataset'] == dataset]
    # plt.scatter(np.log(sdf['pt']), sdf['avg_ratio_n_vertices'])
    Delta0_org = sdf.loc[sdf['pt'].idxmax(), 'Delta0']
    Delta1_org = sdf.loc[sdf['pt'].idxmax(), 'Delta1']
    nn_dist_psD_org = sdf.loc[sdf['pt'].idxmax(), 'nn_dist_psD']
    dmax_psD_org = sdf.loc[sdf['pt'].idxmax(), 'dmax_psD']
    
    ncol = 4
    sdf['log pt'] = np.log(sdf['pt'])
    sdf['Delta0 over org'] = sdf['Delta0']/Delta0_org
    sdf['Delta1 over org'] = sdf['Delta1']/Delta1_org
    sdf['nn_dist_psD over org'] = sdf['nn_dist_psD']/nn_dist_psD_org
    sdf['dmax_psD over org'] = sdf['dmax_psD']/dmax_psD_org

    sdf['local global ratio'] = sdf['nn_dist_psD']/sdf['dmax_psD']
    fig, ax = plt.subplots(1,ncol, figsize=(6*ncol,6))
    T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='dmax_psD', ax=ax[0])
    # T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='Delta0 over org', ax=ax[0])
    # T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='Delta1 over org', ax=ax[0])
    T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='nn_dist_psD', ax=ax[1])
    # T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='avg_ratio_n_vertices', ax=ax[1])
    T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='nn_dist_psD over org', ax=ax[2], color_mean='red')
    T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='dmax_psD over org', ax=ax[2])
    # T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='local global ratio', ax=ax[2])
    T.pl.plot_tradeoff(sdf, xcol='log pt', ycol='l1', ax=ax[3])
    # ax[1].scatter(np.log(sdf['pt']), sdf['avg_ratio_n_vertices'])
    # ax[2].scatter(np.log(sdf['pt']), sdf['nn_dist_psD']/sdf['dmax_psD'])
    # ax[3].scatter(np.log(sdf['pt']), sdf['l1'])
    # plt.scatter(sdf['avg_ratio_n_vertices'], sdf['dmax_psD'])
    plt.suptitle(dataset)
    plt.show()
```
