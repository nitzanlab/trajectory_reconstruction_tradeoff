```python
datasets = [
    # 'linear_rep0',
    'hayashi', 
            # 'dendritic', 
            'beta'
            ]
```

```python
pc_low = 0.03
pc_high = 0.84
```

```python
exp_desc = 'mypseudo'
L_tradeoff_dict = T.io.read_output(datasets, outdir=outdir, sample='exp', exp_desc=exp_desc)
```

```python
for traj_desc, L_tradeoff in L_tradeoff_dict.items():
    assert(L_tradeoff['B'].unique().shape[0] == 1)
    print(L_tradeoff['B'].unique())
```

```python
# dataset = 'hayashi'
# L_tradeoff = L_tradeoff_dict[dataset]
err_fit = 'l1'

# Plots
for traj_desc, L_tradeoff in L_tradeoff_dict.items():
    print(traj_desc)
    res = L_tradeoff
    pc_opt = L_tradeoff.groupby('pc')[err_fit].median().idxmin()
    Pcs_h = [pc_low, pc_opt, pc_high]
    pl_l1 = T.pl.plot_tradeoff_dw(res, Pcs_h,ycol='l1', ylabel='Reconstruction error')
    pl_exp_corr = T.pl.plot_tradeoff_dw(res, Pcs_h, ycol='exp_corr', ylabel='Expression correlation')
    pl_exp_corr_ord = T.pl.plot_tradeoff_dw(res, Pcs_h, ycol='or_exp_corr', ylabel='Expression correlation (ordered)')
    # pl_exp = T.pl.plot_tradeoff_dw(res, Pcs_h, ycol='norm_diff', ylabel='Expression differences')
    # pl_exp_ord = T.pl.plot_tradeoff_dw(res, Pcs_h, ycol='or_norm_diff', ylabel='Expression differences (ordered)')
    pl_pseudo = T.pl.plot_tradeoff_dw(res, Pcs_h, ycol='dpt_corr', ylabel='Pseudotime correlation')

    pl_l1.display()
    pl_exp_corr.display()
    pl_exp_corr_ord.display()
    # pl_exp.display()
    # pl_exp_ord.display()
    pl_pseudo.display()
    if to_save:
        save(pl_l1, os.path.join(outdir, '%s_l1.png' % traj_desc),  scale_factor=2.0)
        save(pl_exp_corr, os.path.join(outdir, '%s_exp.png' % traj_desc),  scale_factor=2.0)
        save(pl_exp_corr_ord, os.path.join(outdir, '%s_exp_ord.png' % traj_desc),  scale_factor=2.0)
        save(pl_pseudo, os.path.join(outdir, '%s_pseudo.png' % traj_desc),  scale_factor=2.0)
```
