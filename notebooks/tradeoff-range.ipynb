```python
def predict_curve(model_cells, model_reads, L_tradeoff, ):
    if 'log pc' not in L_tradeoff.columns:
        L_tradeoff['log pc'] = np.log(L_tradeoff['pc'])
    if 'log pt' not in L_tradeoff.columns:
        L_tradeoff['log pt'] = np.log(L_tradeoff['pt'])

    l1_min = 0.05
    return np.maximum(l1_min, model_cells.predict(L_tradeoff[['log pc']])) +  np.maximum(l1_min, model_reads.predict(L_tradeoff[['log pt']]))
    # return np.maximum(model_cells.predict(L_tradeoff[['log pc']]), model_reads.predict(L_tradeoff[['log pt']]))

    
```
